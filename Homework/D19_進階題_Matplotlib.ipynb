```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#題目 : 將資料夾中boston.csv讀進來，並用圖表分析欄位。
#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?
df1=pd.read_csv('boston.csv')
df1.boxplot()

Median=df1.median().between(300, 400)
print(Median)
print("由圖可以看到B與TAX欄位的中位數位於300-400")
```

```python
#2.畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?
x=df1.NOX
y=df1.DIS
plt.scatter(x,y, c='g', alpha=0.8, s=10)
plt.show()
#由圖可知此兩欄位呈現反比的關係
```
