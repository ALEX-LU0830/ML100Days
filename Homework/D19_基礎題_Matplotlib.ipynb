```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 2*np.pi, 0.01)
y_cos=np.cos(x)

plt.plot(x, y_cos)
plt.title("y=cos(x)", size=15)
plt.savefig("cos(x).png",dpi=300,format="png")
```

```python
n=1024
X=np.random.normal(0,1,n)
Y=np.random.normal(0,1,n)
T=np.arctan2(Y,X)
plt.scatter(X,Y, c=T, s=80, alpha=0.5)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()
```
