```python
import numpy as np
import pandas as pd
```

```python
## 1. 請接下列資料依照指定規則做合併
## 依照fruit做合併 
## 依照Index做合併
df1 = pd.DataFrame({
    'fruit': ['apple', 'banana', 'orange'] * 3,
    'weight': ['high', 'medium', 'low'] * 3,
    'price': np.random.randint(0, 15, 9)
})

df2 = pd.DataFrame({
    'fruit': ['apple', 'orange', 'pine'] * 2,
    'weight': ['high', 'low'] * 3,
    'price': np.random.randint(0, 15, 6)
})
print(df1)
print(df2)
```

```python
pd.merge(df1,df2, on='fruit')
```

```python
pd.merge(df1,df2, on='index')
```

```python
##2. [簡答題] 承上題，請問為什麼依照 merge 合併後有些資料不見了？
因為df1裡面有banana但沒有pine ，但是df2中有pine但沒有banana水果，因此merge的時候只會找到apple跟orange這兩項，導致df1中banana資料不見
以及df2中pine資料不見,修正方法為使用merge中的how=outer來做修正，使banana and pine 項可以跑出來。
```

```python
pd.merge(df1,df2, on='fruit', how='outer')
```

```python
##3. [簡答題] 承上題，請問為什麼依照 index 合併會發生錯誤？請用程式解決。
在使用index做合併時，因為fruit, weight, price這些項目在兩個DataFrame有重複到，導致上述的程式碼發生錯誤，因此需要將column name
改為_df1, _df2來做區別。
```

```python
pd.merge(df1,df2, how='outer', left_index=True, right_index=True ,suffixes=["_df1","_df2"])
```
